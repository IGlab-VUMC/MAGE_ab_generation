```python
import pandas as pd
import numpy as np
```

```python
seqs = pd.read_excel('41467_2018_8165_MOESM4_ESM.xlsx')
```

```python
seqs = seqs[['mAb Name', 'VDJ SEQUENCE']]
seqs['VDJ SEQUENCE'] = seqs['VDJ SEQUENCE'].str.upper()
```

```python
def save_as_fasta(df, filename):
    with open(filename, 'w') as file:
        for index, row in df.iterrows():
            header = f">{row['mAb Name']}\n"
            sequence = f"{row['VDJ SEQUENCE']}\n"
            file.write(header)
            file.write(sequence)
```

```python
save_as_fasta(seqs, 'liu_flu_nuc_seqs_24-01-04.fasta')
```
